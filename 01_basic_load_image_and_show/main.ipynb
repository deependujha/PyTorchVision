## Load image from filesystem using PyTorch and have it in the tensor format ðŸ˜Ž

#### Note:
- `PyTorch keeps images in (C * H * W) {channel, Height, Width}`.
- `But, numpy or matplotlib or many other prefer (H * W * C).`
- PyTorch does so for performance optimizations.
- If using batch of images, then the shape in pytorch is: (N * C * H * W), where N = no. of images in batch

```python
import torch
import numpy as np
from torchvision.io import read_image
```

## Load image and convert to tensor

```python
path_to_image = "data/astronaut.jpg"

my_img = read_image(path_to_image)

# pytorch keeps images in shape (channel * height * width)
print(f"{type(my_img) = } \n{my_img.shape = }")
```

## Visualize image using `matplotlib's imshow()`

```python
import matplotlib.pyplot as plt

# since, imshow() requires images in (H * W * C),
# use pytorch's permute function (https://pytorch.org/docs/stable/generated/torch.permute.html)
plt.imshow(my_img.permute(1, 2, 0).numpy())
plt.show()
```

## Save image tensor to the filesystem ðŸ“‚

```python
from torchvision.io import write_jpeg

write_jpeg(my_img, "data/my_new_astronaut.jpg", quality=90)
```
